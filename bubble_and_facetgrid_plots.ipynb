```python
#libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("patients.csv")
```

```python
data.head()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="weight", y="bmi", data=data)
plt.title("Body Mass Index against weight");
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x="weight", y="bmi", hue="assigned_sex", data=data)
plt.title("Body Mass Index against weight and height");
```

```python
# Facetgrid of assigned_sex and height
grid = sns.FacetGrid(data,col="assigned_sex")
grid.map(plt.hist, "height");
```

```python
# Facetgrid of assigned_sex and weight
grid = sns.FacetGrid(data,col="assigned_sex")
grid.map(plt.hist, "weight");
```
