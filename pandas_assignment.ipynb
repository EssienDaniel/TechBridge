### Assignment
From the application_dat.csv get the month total loan amount requested, and also the month with the overall total loan requests for  the period between june-2018 to december 2018
also get the monthly total loan default in days for each, and also the month with the overall total highest default for  the same period.

```python
# 1. From the application_dat.csv get the month total loan amount requested

# Each month total loan requested all through the period
# Created a new column called month and use a pandas method called dt.month_name()
# To change the datetime to months

data["months"] = data["appilcationDate"].dt.month_name()
month_total = data.groupby("months").agg({"LoanAmount":"sum"})
month_total
```

```python
# 2. The month with the overall total loan requests for the period between june-2018 to december 2018
# Monthly total loan requested between June nad December 2018
data_2018 = data.loc[(data["appilcationDate"].dt.year == 2018)&(data["appilcationDate"].dt.month_name().isin(["June","July","August","September","October","November","December"]))]
data_2018
```

```python
data_2018["months"].value_counts()
```

```python
month_max=data_2018.groupby("months").agg({"LoanAmount":"sum"})
month_max.sort_values(["LoanAmount"])
```

```python
# month with maximum
month_max[month_max.LoanAmount==month_max.LoanAmount.max()]
```

```python
# monthly total default
max_default =data_2018[data_2018["dateDifference"].dt.days > 3].groupby("months").agg({"LoanAmount":"sum"})
max_default = max_default.sort_values(by=["LoanAmount"])
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(max_default.index, max_default["LoanAmount"], label="Loan Amount")
plt.xlabel("Month")
plt.ylabel("Total amount defaulted")
plt.title("Total default amount for the year 2018 between June and December")
```

```python
plt.bar(max_default.index, max_default["LoanAmount"], label="Loan Amount")
plt.xlabel("Month")
plt.ylabel("Total amount defaulted")
plt.title("Total default amount for the year 2018 between June and December")
```

```python
# The month with maximum default
max_default[max_default.LoanAmount == max_default.LoanAmount.max()]
```
